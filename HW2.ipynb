```python
# Game: rock paper scissors
# A function to play rock paper scissors with computer
# Requirements:
# - rule: paper > rock, ...
# - computer randomly choose one
# - player input one choice
# - get result from rule, win / lose / draw
# - play 3 rounds of each game, win 2 to win the game
# - if draw, replay that round

import random

game_set = 3
win_count = 0
lose_count = 0

while True:
  player_choice = input("Please input your choice: rock, paper, scissors")
  computer_choice = random.choice(["rock", "paper", "scissors"])
  print(f"Player choice is: {player_choice}")
  print(f"Compuer choice is: {computer_choice}")
  if player_choice == computer_choice:
      print(f"Both players choice are {player_choice}. Draw.")
  elif player_choice == "rock":
      game_set = game_set - 1
      if computer_choice == "scissors":
          win_count = win_count + 1
          print(f"rock > scissors. You win {win_count} times.")
          if win_count == 2:
              break
      else:
          lose_count = lose_count + 1
          print(f"paper > rock. You lose {lose_count} times.")
          if lose_count == 2:
              break
  elif player_choice == "paper":
      game_set = game_set - 1
      if computer_choice == "rock":
          win_count = win_count + 1
          print(f"paper > rock. You win {win_count} times.")
          if win_count == 2:
              break
      else:
          lose_count = lose_count + 1
          print(f"scissors > paper. You lose {lose_count} times.")
          if lose_count == 2:
              break
  elif player_choice == "scissors":
      game_set = game_set - 1
      if computer_choice == "paper":
          win_count = win_count + 1
          print(f"scissors > paper. You win {win_count} times.")
          if win_count == 2:
              break
      else:
          lose_count = lose_count + 1
          print(f"rock > scissors. You lose {lose_count} times.")
          if lose_count == 2:
              break
  if game_set==0:
      break
```
